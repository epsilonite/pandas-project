```python
# Dependencies
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import hvplot.pandas
import fns
```

```python
yr = 2023
y = 2022
```

```python
gdf = gpd.read_file(f'outputs/geojson/clean/ca_fire_{yr}.geojson')
```

```python
gdf['station'] = pd.read_csv(f'outputs/fire/{yr}/stns_{y}.csv')[f'STN{y}']
```

```python
gdf
```

```python
acres = gdf.groupby('station')['GIS_ACRES'].sum().sort_values()
acres.tail(20)
```

```python
counts = gdf.groupby('station')['AGENCY'].count().sort_values()
counts.tail(20)
```

```python
acres = gdf.groupby('station')['GIS_ACRES'].sum().sort_values()
counts = gdf.groupby('station')['AGENCY'].count().sort_values()
stns = set(counts.tail(20).index) & set(acres.tail(20).index)
# stns = stns | set(counts.tail(3).index)
# stns = stns | set(acres.tail(10).index)
stns = list(stns)
stns
```

```python
stn = stns[3]
```

```python
gdf_0 = gdf[gdf['station']==stn].reset_index(drop=True).drop(columns=['station']).set_crs(crs=4326)
gdf_0 = gdf_0.rename(columns={'ALARM_DATE':'start','CONT_DATE':'end','CAUSE_ID':'cause_id','CAUSE':'cause',
                              'GIS_ACRES':'acres','AGENCY':'agency','FIRE_NAME':'name'})
gdf_0
```

```python
stn_data = pd.read_csv(f'outputs/prcp/{y}/CA_{y}_coords.csv').set_index('station').loc[stn]
```

```python
gdf_0.loc[gdf_0['name'].count(),:] = {'start':f'{yr-1}-07','end':f'{yr}-07','cause_id':'','cause':'Weather Station','acres':float(stn_data['elev']),'agency':stn,
                   'name':stn_data['name'],'geometry':fns.polycircle(stn_data['lon'],stn_data['lat'],fns.box_radius(gdf_0.total_bounds)/100)}
```

```python
gdf_0
```

```python
gdf_0.hvplot(
    frame_width = 700, frame_height = 700,
    geo = True, coastline = True,
    tiles = "OSM", tiles_opts = {'alpha': 0.1},
    color='cause', alpha=0.6,
    hover_cols=['name','acres']
)
```

```python
dts = pd.date_range(str(yr-1)+"-07-01", str(yr)+"-12-31", freq="ME")
dts = [str(x.date())[0:7] for x in dts]
```

```python
df = pd.read_csv(f'outputs/fire/{yr}/prcp_{yr-1}.csv').drop(columns=['station'])
df = df.join(pd.read_csv(f'outputs/fire/{yr}/prcp_{yr}.csv').drop(columns=['station'])).T
df = df[df.index.isin(dts)].T
df = df.set_index(pd.read_csv(f'outputs/fire/{yr}/stns_{yr}.csv').rename(columns={f'STN{yr}':'station'})['station'])
df
```
