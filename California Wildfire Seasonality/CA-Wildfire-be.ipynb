```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import hvplot.pandas
import geopandas as gpd
from datetime import datetime
```

```python
# Create GeoDataFrame from local geojson file downloaded from CalFire
ca_fire = gpd.read_file('Resources/California_Fire_Perimeters_-7970299229233288981.geojson')
```

```python
# Filter Data by year then output relevant json files to outputs
years = pd.Series(range(2012,2023))
for year in years:
    df = ca_fire[ca_fire['YEAR_']==year]
    df.to_file(f'Outputs/ca_fire_{year}.geojson',driver='GeoJSON',drop_id=True)
```
