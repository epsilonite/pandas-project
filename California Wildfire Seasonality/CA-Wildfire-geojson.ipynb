```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import hvplot.pandas
import geopandas as gpd
from datetime import datetime
```

```python
# Create GeoDataFrame from local geojson file downloaded from CalFire
ca_fire = gpd.read_file('outputs/California_Fire_Perimeters_-7970299229233288981.geojson')
```

```python
# Filter Data by year then output relevant json files to outputs
years = pd.Series(range(2012,2024))
for year in years:
    df = ca_fire[ca_fire['YEAR_']==year]
    df.to_file(f'outputs/ca_fire_{year}.geojson',driver='GeoJSON',drop_id=True)
```

```python
df_2012_2017 = ca_fire[(ca_fire['YEAR_']>=2012)&(ca_fire['YEAR_']<2018)]
df_2012_2017.to_file(f'outputs/ca_fire_2012-2017.geojson',driver='GeoJSON',drop_id=True)
```

```python
df_2018_2023 = ca_fire[ca_fire['YEAR_']>=2018]
df_2018_2023.to_file(f'outputs/ca_fire_2018-2023.geojson',driver='GeoJSON',drop_id=True)
```
