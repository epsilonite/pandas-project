```python
# Dependencies
import requests
import pandas as pd
import time
```

```python
df = pd.DataFrame(columns=['STATION','NAME','LONGITUDE','LATITUDE','ELEVATION'])
df
```

```python
station_df = pd.read_csv('outputs/prcp/CA_stns.csv')
lst = station_df['station'].to_list()
```

```python
year = 2023
dts = pd.date_range(str(year) + "-01-01", str(year) + "-12-31", freq="D")
dts_chk = []
```

```python
# Save config information.
url = "https://www.ncei.noaa.gov/access/services/data/v1?"
# Build default Params
params = {'dataset':'daily-summaries',
          'station':'',
          'startDate':'',
          'endDate':'',
          'dataTypes':'PRCP',
          'format':'json',
          'includeAttributes':0,
          'includeStationName':1,
          'includeStationLocation':1,
          'units':'metric'
         }
for stn in lst:
    params['stations'] = stn
    for dt in dts:
        d = str(dt.date())
        params['startDate'] = d
        params['endDate'] = d
        try: 
            response = requests.get(url,params=params).json()
            response_df = pd.DataFrame(response)[['STATION','NAME','LONGITUDE','LATITUDE','ELEVATION']]
            print(response_df)
            df = pd.concat([df,response_df], ignore_index=True)
            break
        except:
            pass
    time.sleep(1)
```
